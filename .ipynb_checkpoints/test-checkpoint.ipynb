```python
import numpy as np
from causallearn.search.ScoreBased.GES import ges
from causallearn.search.ConstraintBased.PC import pc
from causallearn.search.ConstraintBased.FCI import fci
from causallearn.utils.cit import fisherz, chisq, gsq, kci
from causallearn.search.ScoreBased.ExactSearch import bic_exact_search
from causallearn.graph.SHD import SHD
from causallearn.utils.DAG2CPDAG import dag2cpdag
from causallearn.utils.TXT2GeneralGraph import txt2generalgraph
```

### Load the datasets

```python
linear_data_path = "./TestData/data_linear_10.txt"
discrete_data_path = "./TestData/data_discrete_10.txt"
graph_path = "./TestData/graph.10.txt" 

linear_data = np.loadtxt(linear_data_path, skiprows=1)
discrete_data = np.loadtxt(discrete_data_path, skiprows=1)
DAG = txt2generalgraph(graph_path)
CPDAG = dag2cpdag(DAG)
```

#### Manually try PC with different indep_test

```python
# Linear data, only Fisherz works
model = pc(linear_data, alpha=0.05, indep_test="fisherz")
shd = SHD(CPDAG, model.G)
print("SHD score: %.2f" % shd.get_shd())
```

```python
# Discrete data
model = pc(discrete_data, alpha=0.05, indep_test="fisherz")
shd = SHD(CPDAG, model.G)
print("SHD score: %.2f" % shd.get_shd())

model = pc(discrete_data, alpha=0.05, indep_test="chisq")
shd = SHD(CPDAG, model.G)
print("SHD score: %.2f" % shd.get_shd())

model = pc(discrete_data, alpha=0.05, indep_test="gsq")
shd = SHD(CPDAG, model.G)
print("SHD score: %.2f" % shd.get_shd())
```

#### Manually try PC with different uc_rule

```python
# uc_sepset
model = pc(linear_data, alpha=0.05, indep_test="fisherz", uc_rule=0)
shd = SHD(CPDAG, model.G)
print("SHD score: %.2f" % shd.get_shd())

# maxP
model = pc(linear_data, alpha=0.05, indep_test="fisherz", uc_rule=1)
shd = SHD(CPDAG, model.G)
print("SHD score: %.2f" % shd.get_shd())

# definiteMaxP
model = pc(linear_data, alpha=0.05, indep_test="fisherz", uc_rule=2)
shd = SHD(CPDAG, model.G)
print("SHD score: %.2f" % shd.get_shd())
```

#### Manually try PC with different uc_priority

```python
# default in uc_rule
model = pc(linear_data, alpha=0.05, indep_test="fisherz", uc_rule=2, uc_priority=-1)
shd = SHD(CPDAG, model.G)
print("SHD score: %.2f" % shd.get_shd())

# prioritize existing colliders
model = pc(linear_data, alpha=0.05, indep_test="fisherz", uc_rule=2, uc_priority=2)
shd = SHD(CPDAG, model.G)
print("SHD score: %.2f" % shd.get_shd())

# prioritize stronger colliders
model = pc(linear_data, alpha=0.05, indep_test="fisherz", uc_rule=2, uc_priority=3)
shd = SHD(CPDAG, model.G)
print("SHD score: %.2f" % shd.get_shd())

# prioritize stronger* colliders
model = pc(linear_data, alpha=0.05, indep_test="fisherz", uc_rule=2, uc_priority=4)
shd = SHD(CPDAG, model.G)
print("SHD score: %.2f" % shd.get_shd())
```

#### HPO: PC 

```python
from ConfigSpace import ConfigurationSpace
from ConfigSpace.hyperparameters import UniformIntegerHyperparameter, CategoricalHyperparameter, UniformFloatHyperparameter
from smac.facade.smac_bb_facade import SMAC4BB
from smac.facade.smac_hpo_facade import SMAC4HPO
from smac.facade.smac_ac_facade import SMAC4AC
from smac.scenario.scenario import Scenario
```

```python
def train_pc_algorithm(config):
    """ 
    Objective function (minimizes)
    Input: configuration
    Output: SHD value
    
    """
    model = pc(
        discrete_data, 
        alpha=0.05, 
        indep_test=config["indep_test"], 
        uc_rule=config["uc_rule"], 
        uc_priority=config["uc_priority"]
    )
    shd = SHD(CPDAG, model.G)
    return shd.get_shd()
```

```python
# Define configuration space
configspace = ConfigurationSpace()

# Different indep_test only works with discrete data (for linear data, only fihsherz works)
indep_test = CategoricalHyperparameter("indep_test", ["fisherz", "chisq", "gsq"], default_value="fisherz")
configspace.add_hyperparameter(indep_test)
uc_rule = CategoricalHyperparameter("uc_rule", [0, 1, 2], default_value=0)
configspace.add_hyperparameter(uc_rule)
uc_priority = CategoricalHyperparameter("uc_priority", [-1, 2, 3, 4], default_value=2)
configspace.add_hyperparameter(uc_priority)
# print("Sample two configurations: \n", configspace.sample_configuration(2))

scenario = Scenario({
    "run_obj": "quality",   # optimize objective function (either quality or runtime)
    "runcount-limit": 6,    # max number of function evaluations
    "cs": configspace, 
    "deterministic": False,
})

def_value = train_pc_algorithm(configspace.get_default_configuration())
print("Default Value: %.2f" % def_value)

smac = SMAC4HPO(
    scenario=scenario, 
    rng=np.random.RandomState(42),
    tae_runner=train_pc_algorithm,
)

best_found_config = smac.optimize()
best_value = train_pc_algorithm(best_found_config)
print("Best found configuration: \n", best_found_config)
print("Optimized Value: %.2f" % best_value)

# Show runhistory
# rh = smac.get_runhistory()
# for (config_id, instance_id, seed, budget), (cost, time, status, starttime, endtime, additional_info) in rh.data.items():
#     config = rh.ids_config[config_id]
#     print(config)
```

#### Manually try FCI with different independence_test_method

```python
# Linear data
G, edges = fci(linear_data, independence_test_method="fisherz")
shd = SHD(CPDAG, G)
print("SHD score: %.2f" % shd.get_shd())

# Discrete data
G, edges = fci(discrete_data, independence_test_method="fisherz")
shd = SHD(CPDAG, G)
print("SHD score: %.2f" % shd.get_shd())

G, edges = fci(discrete_data, independence_test_method="chisq")
shd = SHD(CPDAG, G)
print("SHD score: %.2f" % shd.get_shd())

G, edges = fci(discrete_data, independence_test_method="gsq")
shd = SHD(CPDAG, G)
print("SHD score: %.2f" % shd.get_shd())
```

#### HPO: FCI

```python
def train_fci_algorithm(config):
    """ 
    Objective function (minimizes)
    Input: configuration
    Output: SHD value
    
    """
    G, edges = fci(
        discrete_data, 
        alpha=0.05, 
        independence_test_method=config["independence_test_method"]
    )
    shd = SHD(CPDAG, G)
    return shd.get_shd()
```

```python
# Define configuration space
configspace = ConfigurationSpace()

# Different indep_test only works with discrete data (for linear data, only fihsherz works)
independence_test_method = CategoricalHyperparameter("independence_test_method", ["fisherz", "chisq", "gsq"], default_value="fisherz")
configspace.add_hyperparameter(independence_test_method)
# print("Sample two configurations: \n", configspace.sample_configuration(2))

scenario = Scenario({
    "run_obj": "quality",   # optimize objective function (either quality or runtime)
    "runcount-limit": 10,    # max number of function evaluations
    "cs": configspace, 
    "deterministic": False,
})

def_value = train_fci_algorithm(configspace.get_default_configuration())
print("Default Value: %.2f" % def_value)

smac = SMAC4HPO(
    scenario=scenario, 
    rng=np.random.RandomState(42),
    tae_runner=train_fci_algorithm,
)

best_found_config = smac.optimize()
best_value = train_fci_algorithm(best_found_config)
print("Best found configuration: \n", best_found_config)
print("Optimized Value: %.2f" % best_value)

# Show runhistory
# rh = smac.get_runhistory()
# for (config_id, instance_id, seed, budget), (cost, time, status, starttime, endtime, additional_info) in rh.data.items():
#     config = rh.ids_config[config_id]
#     print(config)
```

#### HPO: GES

```python
def train_ges_algorithm(config):
    """ 
    Objective function (minimizes)
    Input: configuration
    Output: SHD value
    
    """
    model = ges(
        discrete_data, 
        score_func=config["score_func"]
    )
    shd = SHD(CPDAG, model['G'])
    return shd.get_shd()
```

```python
# Define configuration space
configspace = ConfigurationSpace()

# Different indep_test only works with discrete data (for linear data, only fihsherz works)
score_func = CategoricalHyperparameter("score_func", ["local_score_BIC", "local_score_BDeu"], default_value="local_score_BIC")
configspace.add_hyperparameter(score_func)
# print("Sample two configurations: \n", configspace.sample_configuration(2))

scenario = Scenario({
    "run_obj": "quality",   # optimize objective function (either quality or runtime)
    "runcount-limit": 2,    # max number of function evaluations
    "cs": configspace, 
#     "deterministic": False,
})

def_value = train_ges_algorithm(configspace.get_default_configuration())
print("Default Value: %.2f" % def_value)

smac = SMAC4HPO(
    scenario=scenario, 
    rng=np.random.RandomState(42),
    tae_runner=train_ges_algorithm,
)

best_found_config = smac.optimize()
best_value = train_ges_algorithm(best_found_config)
print("Best found configuration: \n", best_found_config)
print("Optimized Value: %.2f" % best_value)

# Show runhistory
# rh = smac.get_runhistory()
# for (config_id, instance_id, seed, budget), (cost, time, status, starttime, endtime, additional_info) in rh.data.items():
#     config = rh.ids_config[config_id]
#     print(config)
```
